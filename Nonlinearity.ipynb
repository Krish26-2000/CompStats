# Nonlinearity 

So far: mostly focused on linear models.


Linear models: simple to describe, implement, and interpret. Inference is (relatively) simple.

However: can have significant limitations in terms of predictive power

Ridge regression, the lasso, principal components regression, and other techniques: improvement is obtained by reducing the complexity of the linear model, and hence the variance of the estimates. But: model is still linear model.


- polynomial regression 
- step functions
- splines  




## Polynomial Regression


Replace the linear regression model

$$
y_{i}=\beta_{0}+\beta_{1} x_{i}+\epsilon_{i}
$$

with a polynomial function

\begin{equation}\label{eq:}
y_{i}=\beta_{0}+\beta_{1} x_{i}+\beta_{2} x_{i}^{2}+\beta_{3} x_{i}^{3}+\ldots+\beta_{d} x_{i}^{d}+\epsilon_{i}
\end{equation}

For large regression
enough degree d, a polynomial regression allows us to produce an extremely
non-linear curve.
